***1 will indicate a perfect positive relationship (value increases when feature's value also increases)***

***R² score > 0.5 indicates strong correlation, feature will be selected, R² < 0.5 indicates weak correlation, feature will not be selected***

### Measuring the Relationship between Gameplay Performance and KDA

```python
# Prepare the data
X = df['Gameplay Performance (100%)'].values.reshape(-1, 1)
y = df['Average KDA'].values 

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict using the model
y_pred = model.predict(X)

# Calculate R^2 score
r2 = r2_score(y, y_pred)

# Plotting the regression
plt.figure(figsize=(10, 6))
sns.regplot(x='Gameplay Performance (100%)', y='Average KDA', data=df, scatter_kws={'s': 100}, line_kws={'color': 'red'})

# Customize the plot
plt.title("Regression of Gameplay Performance vs Average KDA", fontsize=14)
plt.xlabel("Gameplay Performance (100%)", fontsize=12)
plt.ylabel("Average KDA", fontsize=12)

# Annotate R^2 score
plt.text(0.1, 0.9, f'$R^2$: {r2:.4f}', fontsize=12, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))

# Display the plot
plt.show()
```

## Model Buildings

##### As the ranks of the game were a categorical variable to be considered, the current project will focus mainly on utilizing classification models for predicting students' rank

4 different types of machine learning models will be tested to determine the best models to be utilized for further prediction:
1. Logistic Regression (LR)
2. Random Forest Classifier (RFC)
3. Support Vector Classifier (SVC)
4. K-Nearest Neighbors (KNN)

GridSearchCV will be performed to tune the best hyperparameters for each model to ensure the best performance were brought out by each models.

The performance of each models will be tested by 4 types of scores:
1. R² score
2. Jaccard Index Score
3. Metrics Accuracy Score
4. F1 Score

### Preparing and Filtering Data

```python
# Features and target preparation
X = df['Gameplay Performance (100%)'].values.reshape(-1, 1)  # Reshaping for single feature compatibility
y = df['Rank']

# Splitting data: Q1 to Q3 for training and Q4 for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, shuffle=False)

# Display dataset sizes
print("Size of training dataset:", X_train.shape, y_train.shape)
print("Size of testing dataset:", X_test.shape, y_test.shape)
```

```python
# Model definitions and parameter grids
models = {
    "Logistic Regression": {
        "model": LogisticRegression(max_iter=1000),
        "params": {'C': [0.1, 1, 10], 'solver': ['newton-cg', 'lbfgs', 'liblinear']}
    },
    "SVC": {
        "model": SVC(),
        "params": {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto']}
    },
    "Random Forest": {
        "model": RandomForestClassifier(),
        "params": {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
    },
    "KNN": {
        "model": KNeighborsClassifier(),
        "params": {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
    }
}

# Evaluating all models
best_model = None
best_accuracy = 0

for model_name, config in models.items():
    grid = GridSearchCV(config["model"], config["params"], cv=3)
    grid.fit(X_train, y_train)
    y_pred = grid.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    r2 = r2_score(y_test, y_pred)
    jaccard = jaccard_score(y_test, y_pred, average='weighted')

    print(f"Model: {model_name}")
    print(f"Best Parameters: {grid.best_params_}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"R² Score: {r2:.4f}")
    print(f"Jaccard Index: {jaccard:.4f}")
    print("-" * 40)
    
    # Display the best model
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = (model_name, grid.best_estimator_)

print(f"The best model is: {best_model[0]} with an accuracy of {best_accuracy:.4f}")
```

***Based on the output above, it seems that all 4 models were showing low accuracy score to predict students' rank over the next quarter. This can be due to 3 factors:***
1. As the rank of the game were typically based on the game's personal build ranking system, it is hard to predict students' next rank as we have no information on how the ranking system would promote or demote a students' individual ranks
2. The sample size of this dataset were relatively small, there was not enough evidence for gameplay performance to predict students' rank with a sample size as small as this.
3. Adding on, given that gameplay performance were students' performance in terms of gameplay measured by the coach, it is skewed differently as compared to how the game's ranking system would perceive the students' gameplay performance, since only the system itself understood on the students' gameplay performance in regards to their ranks.

**In that case, it is difficult to utilize gameplay performance to predict students' rank over the next quarter. That said, we can try a different approach.**

### Predicting Students' Ability to achieve their Rank goals by assessing Current Rank, Rank Goals, and their Growth in Rank

Since students' improvements were able to be monitored by their ranks' improvements, yet it was hard to predict their rank over the next quarter, we can predict whether students would be able to achieve their rank goals for the next quarter. 

However, since the rank goals were established by the agreement between the coach and the students, some of the rank goals may seem slightly different. 

Besides, since all students have managed to increase their rank placement at some point, we can use the gap between students' rank growth as a predictor for whether students may achieve their rank goals or not. Below table indicates the rank goals system and how it works throughout each quarter:

Quarters | Rank Goal
---|---
Q1 | Coach Assigned Rank Goal of Q1
Q2 | Students' Final Rank in Q1 + 3 ranks *(eg: Silver 1 in Q1, Rank goal in Q2 is Gold 1, 3 ranks above Silver 1)*
Q3 | Students' Final rank in Q2 + 3 ranks
Q4 | Student and Coach agreed rank goals *(Established in the first class)*

### Adding Relevant Datas into the Dataset for further predictive Analysis

For this, the relevant data must first be added into the dataset. The new columns will consist of "Rank Goals", "Achieve", and "Rank Growth"

##### Next, the dataset will be split into training sets and testing sets, where Q1 to Q3 data will be the training set, with Q4 data being the testing set.

##### the 4 models will then be implemented to test the accuracy of predictions to locate the best model across all 4 models, Confusion Matrix will be displayed for better visualizations

```python
features = ["Rank Growth","Rank Goals","Rank"]
target = "Achieve"

train_data = df[df["Quarter"] != "Q4"]
test_data = df[df["Quarter"] == "Q4"]

X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]
y_test = test_data[target]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

model_results = {}

# plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred)
    ax = plt.subplot()
    sns.heatmap(cm, annot=True, fmt='d', ax=ax) 
    ax.set_xlabel('Predicted Labels', fontsize=12)
    ax.set_ylabel('True Labels', fontsize=12)
    ax.set_title(title, fontsize=15)
    ax.xaxis.set_ticklabels(['Did Not Achieve', 'Achieved'])
    ax.yaxis.set_ticklabels(['Did Not Achieve', 'Achieved'])
    plt.show()

# Training and evaluating models
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring="accuracy")
    y_pred = model.predict(X_test_scaled)
    test_accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred)
    
    model_results[name] = {
        "Model": model,
        "CV Scores": cv_scores,
        "Mean CV Accuracy": np.mean(cv_scores),
        "Test Accuracy": test_accuracy,
        "Classification Report": classification_rep,
        "Confusion Matrix": confusion_matrix(y_test, y_pred)
    }
    
    # Display classification report and confusion matrix for each model
    print(f"\nModel: {name}")
    print("Classification Report:\n", classification_rep)
    plot_confusion_matrix(y_test, y_pred, title=f"{name} Confusion Matrix")

# Find the best model based on test accuracy
best_model = max(model_results.items(), key=lambda x: x[1]["Test Accuracy"])
print(f"\nThe best model is {best_model[0]} with Test Accuracy: {best_model[1]['Test Accuracy']:.4f}")

# Summary of all models
print("\nSummary of all models:")
for model_name, result in model_results.items():
    print(f"\nModel: {model_name}")
    print(f"Cross-validation Scores: {result['CV Scores']}")
    print(f"Mean CV Accuracy: {result['Mean CV Accuracy']:.4f}")
    print(f"Test Accuracy: {result['Test Accuracy']:.4f}")
```
