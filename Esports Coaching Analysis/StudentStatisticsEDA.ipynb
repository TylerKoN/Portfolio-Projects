Students statistics were recorded at the end of each quarters to monitor performances of students and observed for improvements. Datas are collected via Coach's Evaluation and Students' Game Tracker in terms of mechanical aspect of the statistics, mainly KDA and Highest Rank achieved during the Quarter. The table below displays the Variables and each of its intepretation methods within the DataFrame

#### Variables Table

Variables | Variable Intepretation methods
----------|--------------------------------------------
Student | Student's Name.
Quarter | The quarters of the year 2024, with Q1 being the first Quarter, Q2 as second Quarter and so on. Each Quarters are measured to be 10 weeks, or 2.5 months.
Rank | Student's Highest Achived Rank throughout the Quarter, (*A more specific Table will be provided below in* ***External Factors***)
Gameplay Performance (100%) | Students' Gameplay performance recorded and evaluated by the coach, assesing on students' game's knowledge, mechanical aspect, and problem-solving within the game's nature 
Average KDA | The Average value of students' general in-game "Kills, Deaths, Assists" value throughout the Quarter, with the calculation format being ***"(Total Kills + Total Assists)/Total Deaths"***.
Quarter Leaderboard Rank | Students' Final class-Leaderboard Rank at the end of each Quarter.
Discipline Score (25%) | Students' in-game discipline score that assessed on students' performance in the game, to measure whether student remain calm and discipline or always following his ego. 
Communication Score (25%) | Students' ability to communicate information, messages, and callouts with his teammate, this also assess the students' communicating behaviours in the team such as being polite, or being rude. 
Teamwork Score (25%) | Students' ability to adapt and work with others in different scenarios, this also assesses students' in-game ability of supporting others, playing together as a team, and cheering each others.
Leadership Score (15%) | Students' ability to lead the team in the game or in physical state, this also assesses students' confidence.
Sportsmanship Score (10%) | Coach's evaluation of students' behaviours in terms of sportsmanship, in which whether the student respects the opponents, students' level of professionalism in and out the team, and students' commitment in his trainings. 
Final Score (100%) | The sum value of students' Discipline, Communication, Teamwork, Leadership, and Sportsmanship score throughout the Quarter.

## Students' Final Score Throughout the Quarter

```python
students = df['Student'].unique()
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

plt.figure(figsize=(15, 10))

# Creating a subplot
for idx, student in enumerate(students):
    student_data = df[df['Student'] == student].copy()
    final_scores = student_data['Final Score (100%)'].tolist()
    
    # Special handling for Joe's data
    if student == 'Joe':
        final_scores = [np.nan] + final_scores  # Placeholder NaN for Q1

    if len(final_scores) < len(quarters):
        final_scores += [np.nan] * (len(quarters) - len(final_scores))

    # Create subplot
    plt.subplot(2, 4, idx + 1)  # 2 rows, 4 columns
    plt.plot(quarters, final_scores, linestyle='-', color='red')

    # Find max score and its position
    valid_scores = [(i, score) for i, score in enumerate(final_scores) if not np.isnan(score)]
    if valid_scores:
        max_index, max_score = max(valid_scores, key=lambda x: x[1])
        min_index, min_score = min(valid_scores, key=lambda x: x[1])

        # Place marker and text for max score on the left side of the marker
        plt.scatter(quarters[max_index], max_score, color='purple', zorder=5)
        plt.scatter(quarters[min_index], min_score, color='purple', zorder=5)
        
        # Adjusting text position
        plt.text(quarters[max_index], max_score, f"{int(max_score)}", 
                 horizontalalignment='right', verticalalignment='center', color='black', 
                 fontsize=10, position=(quarters[max_index], max_score + 1.5))
        plt.text(quarters[min_index], min_score, f"{int(min_score)}", 
                 horizontalalignment='center', verticalalignment='center', color='black', 
                 fontsize=10, position=(quarters[min_index], min_score + 1.7))

    # Title and labels
    plt.title(f"{student}'s Final Score")
    plt.xlabel('Quarter')
    plt.ylabel('Final Score')
    plt.xticks(quarters) 
    plt.grid(True)
    plt.ylim(30,100)

plt.tight_layout()
plt.show()
```

The plot above shows clearly on the improvements on every students, where all students have improved to at least 70% and above of their Final Score.

Between the students, Adrian, Jared, Joshua, and Tze Kent have shown the most improvements across all students, with Adrian having the highest score of 95%

```python
students = df['Student'].unique()
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# Create a list to hold the final scores
all_final_scores = []

colormap = cm.tab10  
colors = [colormap(i % 10) for i in range(len(students))]  # Cycle colormap through the first 10 colors

# Loop through each student and prepare the data for plotting
for student in students:
    student_data = df[df['Student'] == student].copy()
    final_scores = student_data['Final Score (100%)'].tolist()

    # Special handling for Joe's data
    if student == 'Joe':
        final_scores = [np.nan] + final_scores  # Placeholder NaN for Q1

    if len(final_scores) < len(quarters):
        final_scores += [np.nan] * (len(quarters) - len(final_scores))

    all_final_scores.append(final_scores)

# Convert the list of final scores into a numpy array
all_final_scores = np.array(all_final_scores)

# Creating the plot
plt.figure(figsize=(12, 8))

# Loop to plot each student's line
for idx, student in enumerate(students):
    # Plotting the line
    plt.plot(quarters, all_final_scores[idx], linestyle='-', label=student, color=colors[idx], linewidth=1.5)

    # Placing marker at the highest point for each student (max value)
    max_score_idx = np.nanargmax(all_final_scores[idx]) 
    plt.plot(quarters[max_score_idx], all_final_scores[idx][max_score_idx], 
             marker='s', color=colors[idx], markersize=5, markeredgecolor='black', markeredgewidth=0.8)

    # Place a marker at the lowest point for each student (min value)
    min_score_idx = np.nanargmin(all_final_scores[idx]) 
    plt.plot(quarters[min_score_idx], all_final_scores[idx][min_score_idx], 
             marker='s', color=colors[idx], markersize=5, markeredgecolor='black', markeredgewidth=0.8)

# Add title and labels
plt.title("Students' Final Scores Throughout the Quarters")
plt.xlabel('Quarters')
plt.ylabel('Final Score')
plt.ylim(20,100)

# Add legend
plt.legend(loc='upper left')

# Display plot
plt.grid(True)
plt.show()
```

Based on the Line Plot above, it is clear that all students have shown significant improvements over the quarters, especially for Adrian, being the highest final scorer in the class.

Joshua and Jared have also displayed incredible improvements based on the graph, where their scores have increased significantly over time and catching up to other players within the classroom.

However, Cayden is the only student who have displayed the least amount of improvements. This can be due to his behaviours in class where his ego often comes into place and causing him to improve significantly slower in terms of discipline, sportsmanship, and communication score.
